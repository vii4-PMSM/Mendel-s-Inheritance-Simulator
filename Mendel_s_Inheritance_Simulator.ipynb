```python
#Step 1 Dad/Mom DNA 23쌍 생성
#e.g. [[DAD1우성, DAD1열성], [DAD2열성, DAD2우성]] - 일단 패스

import random

Dna_w = ["우성", "열성"]

Dad_Dna = [[f"DAD{d+1}a{random.choices(Dna_w, weights = [3, 1])[0]}",
           f"DAD{d+1}b{random.choices(Dna_w, weights = [3, 1])[0]}"]
           for d in range(22)]

sex_pair = ["DadX", "DadY"]
random.shuffle(sex_pair)
Dad_Dna.append(sex_pair)

Mom_Dna = [[f"MOM{d+1}a{random.choices(Dna_w, weights = [3, 1])[0]}",
           f"MOM{d+1}b{random.choices(Dna_w, weights = [3, 1])[0]}"]
           for d in range(22)]
Mom_Dna.append(["MomX", "MomX"])

print("Step 1 Dad/Mom DNA 23쌍 생성")
print("Dad's Dna:", Dad_Dna)
print("Mom's Dna:", Mom_Dna)
print("")
#---------------Step 2 부모 DNA 물려받기---------------
Child_Dna = []
for i in range(23):
  Dad_pick = random.choice(Dad_Dna[i])
  Mom_pick = random.choice(Mom_Dna[i])
  Child_Dna.append([Dad_pick, Mom_pick])

print("Step 2 부모 DNA 물려받기")
print("Child_Dna:", Child_Dna)
print("")
#---------------Step 3 우성/열성 구분하기---------------
Child_Gen = []
for pair in Child_Dna[:22]: #
  Dad_Gene = pair[0]
  Mom_Gene = pair[1]

  if "우성" in Dad_Gene or "우성" in Mom_Gene:
    Child_Gen.append("우성")
  else:
    Child_Gen.append("열성")

print("Step 3 우성/열성 구분하기")
print("Child_DorR:", Child_Gen)
print("")
#---------------Step 4 성별 구분하기---------------
print("Step 4 성별 구분하기")
if "DadX" in Dad_Gene:
  print("Sex/성별:","Female/여성")
else:
  print("Sex/성별:","Male/남성")
```
